```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
df_train = df_train.fillna(0)
features = df_train[['NU_NOTA_LC', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
reg.fit(features, df_train['NU_NOTA_MT'])
```

```python
pd.DataFrame(zip(features.columns, reg.coef_), columns=['features', 'estimatedCoefficients'])
```

```python
reg.predict(features)
```

```python
fig, axes = plt.subplots()
sns.kdeplot(reg.predict(features), shade=True, color='r')
sns.kdeplot(df_train['NU_NOTA_MT'], shade=True, color='g')
```

```python
plt.scatter(features, reg.predict(features), s=3)
plt.plot(features, (features * reg.coef_ + reg.intercept_), color='r')
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(features, df_train['NU_NOTA_MT'], test_size=0.33, random_state=5)
print(X_train) 
print(X_test) 
print(Y_train)
print(Y_test)
```
