```python
features = ['NU_NOTA_LC', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']
df_train[features].corr()
```

```python
df_train = df_train.loc[(df_train['NU_NOTA_LC'].notnull()) & (df_train['NU_NOTA_CH'].notnull()) & (df_train['NU_NOTA_CN'].notnull()) & (df_train['NU_NOTA_MT'].notnull()) & (df_train['NU_NOTA_REDACAO'].notnull())]
```

```python
df_train = df_train[['NU_NOTA_LC', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']]
df_test = df_test[['NU_NOTA_LC', 'NU_NOTA_CH', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']]
```

```python
df_train['NU_NOTA_LC'].fillna(df_train['NU_NOTA_LC'].mean(), inplace=True)
df_train['NU_NOTA_CH'].fillna(df_train['NU_NOTA_CH'].mean(), inplace=True)
df_train['NU_NOTA_CN'].fillna(df_train['NU_NOTA_CN'].mean(), inplace=True)
df_train['NU_NOTA_REDACAO'].fillna(df_train['NU_NOTA_REDACAO'].mean(), inplace=True)
#df_train['TP_DEPENDENCIA_ADM_ESC'].fillna(df_train['TP_DEPENDENCIA_ADM_ESC'].mean(), inplace=True)
df_test['NU_NOTA_LC'].fillna(df_train['NU_NOTA_LC'].mean(), inplace=True)
df_test['NU_NOTA_CH'].fillna(df_train['NU_NOTA_CH'].mean(), inplace=True)
df_test['NU_NOTA_CN'].fillna(df_train['NU_NOTA_CN'].mean(), inplace=True)
df_test['NU_NOTA_REDACAO'].fillna(df_train['NU_NOTA_REDACAO'].mean(), inplace=True)
#df_train['TP_DEPENDENCIA_ADM_ESC'].fillna(df_train['TP_DEPENDENCIA_ADM_ESC'].mean(), inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = preprocessing.StandardScaler().fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_train_scaled.std(axis=1)
```

```python
pipeline = make_pipeline(preprocessing.StandardScaler(), RandomForestRegressor(n_estimators=200, n_jobs=-1, warm_start=True))
```

```python
hyperparameters = {'randomforestregressor__max_features': ['auto', 'sqrt', 'log2'], 'randomforestregressor__max_depth': [None, 5, 3, 1]}
```

```python
clf = GridSearchCV(pipeline, hyperparameters, cv=10)
clf.fit(x_train_scaled, y)
```

```python
print(clf.best_params_)
```

```python
pred = clf.predict(x_test)
from sklearn.preprocessing import normalize
normalize(pred[:, newaxis])
```

```python
df_resposta = pd.DataFrame()
df_resposta['NU_NOTA_MT'] = np.around(np.sort(pred), 2)
```

```python
fig, axes = plt.subplots()
sns.kdeplot(df_resposta['NU_NOTA_MT'], shade=True, color='g')
```
